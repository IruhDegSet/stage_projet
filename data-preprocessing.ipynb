```python
def clean_price(price):
    if isinstance(price, str):
        # Remplacer les virgules par des points et enlever les ?
        price = price.replace(',', '.').replace('?', '')
    try:
        return float(price)
    except ValueError:
        return None

# Appliquer la fonction à la colonne 'Prix'
df8['Prix'] = df8['Prix'].apply(clean_price)
```
